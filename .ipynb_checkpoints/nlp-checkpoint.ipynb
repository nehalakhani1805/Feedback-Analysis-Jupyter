```python
from collections import defaultdict 
```

```python
class Teacher:
    def __init__(self,a):
        self.no_of_p=0
        self.no_of_nt=0
        self.no_of_ng=0
        self.pos=[]
        self.neg=[]
        self.neut=[]  
        self.name=a
        self.p_with_time=defaultdict(int)
        self.nt_with_time=defaultdict(int)
        self.ng_with_time=defaultdict(int)
```

```python
for t in teachers_list:
    t_obj=Teacher(t)
    f2=feedback[feedback.Teacher==t]
    #fa=f2['Review']
    fb=f2['time']
    #print(fb.head())
    #print("done")
type(fb[10])
fb.get_loc(1)
```

```python
#column=['feedback']
#df=pd.read_csv("F://project/Feedback.csv",names=column,header=None)

#fa=feedback['Review']['Teacher'='SK']
individual_teacher_feedback=[]
for t in teachers_list:
    t_obj=Teacher(t)
    f2=feedback[feedback.Teacher==t]
    fa=f2['Review']
    fb=f2[['Review','time']]
    #fa.head()
    f2['Review']=f2['Review'].astype(str)
    f2.dropna(subset=['Review'])
    dict={'positive':0,'negative':0,'neutral':0}
    positive=[]
    negative=[]
    neutral=[]


    lemmatizer = WordNetLemmatizer()
    ps=PorterStemmer()
    #j=0
    for j,i in enumerate(fa):
        #remove punctuation
        tokenizer = RegexpTokenizer(r'\w+')
        tokens = tokenizer.tokenize(i)
        final=' '.join(tokens)


        #remove non english words
        a1=" ".join(w for w in nltk.wordpunct_tokenize(final)if w.lower() in words or not w.isalpha())

        #remove proper nouns
        tokenized2 = nltk.word_tokenize(a1)
        tokenized=[lemmatizer.lemmatize(w) for w in tokenized2]
        #tokenized=[ps.stem(w) for w in tokenized3]
        pos=nltk.tag.pos_tag(tokenized)
        ed=[word for word,tag in pos if tag!='NNP' and tag!='NNPs']#removing proper nouns
        end=' '.join(ed)

        #remove stop words
        en=[i for i in word_tokenize(end.lower()) if i not in stop] 
        final=' '.join(en)

        #using vader classfiers
        sid=SentimentIntensityAnalyzer()
        ss=sid.polarity_scores(final)
        if(0.3<=ss['compound']<=0.5 ):
#             if fb[j] not in t_obj.nt_with_time.keys():
#                 t_obj.nt_with_time[fb[j]]=1
#             else:
            temp=fb.loc[fb['Review']==i,'time'].iloc[0]
            t_obj.nt_with_time[temp]+=1
            dict['neutral']+=1
            t_obj.neut.append(final)
        elif(ss['compound']>0.5):
#             if fb[j] not in t_obj.p_with_time.keys():
#                 t_obj.p_with_time[fb[j]]=1
#             else:
            temp=fb.loc[fb['Review']==i,'time'].iloc[0]
            t_obj.p_with_time[temp]+=1
            dict['positive']+=1
            t_obj.pos.append(final)
        else:
#             if fb[j] not in t_obj.ng_with_time.keys():
#                 t_obj.ng_with_time[fb[j]]=1
#             else:
            temp=fb.loc[fb['Review']==i,'time'].iloc[0]
            t_obj.ng_with_time[temp]+=1
            dict['negative']+=1
            t_obj.neg.append(final)
        #j+=1
    #li=[]
    t_obj.no_of_p=dict['positive']
    t_obj.no_of_nt=dict['neutral']
    t_obj.no_of_ng=dict['negative']
    individual_teacher_feedback.append(t_obj)
labels = 'Positive', 'Neutral', 'Negative'
sizes = [individual_teacher_feedback[0].no_of_p,individual_teacher_feedback[0].no_of_nt,individual_teacher_feedback[0].no_of_ng]
colors = ['gold', 'lightcoral','yellowgreen']
explode = (0.1, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```

```python
#horizontal stack bar graph for comparing teachers
N = 10
p=[]
ng=[]
nt=[]
for i in individual_teacher_feedback:
    p.append(i.no_of_p)
    ng.append(i.no_of_ng)
    nt.append(i.no_of_nt)
#print(p[0],nt[0],ng[0])

ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence
#print(ind)
p1 = plt.barh(ind, p, width,color='g')
p2 = plt.barh(ind, nt, width,left=p,color='y')
p3 = plt.barh(ind, ng, width,left=np.array(p)+np.array(nt),color='r')

plt.xlabel('Feedback Scores')
plt.title('Scores by sentiment')
plt.yticks(ind, teachers_list)
plt.xticks(np.arange(0, 4001, 500))
plt.legend((p1[0], p2[0],p3[0]), ('Positive','Neutral','Negative'))

plt.show()
```
